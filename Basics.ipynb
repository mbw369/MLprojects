```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
yes_no={"Yes":True, "No":False}
star_wars["Have you seen any of the 6 films in the \
Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to \
be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
movie_rename = star_wars.columns[3:9].tolist()
#print (movie_rename)
mapping_Nan_movies={}
for i,movie_name in enumerate(movie_rename, 1):
   
    mapping_Nan_movies[movie_name]="seen" + str(i)
print(mapping_Nan_movies)
star_wars = star_wars.rename(columns=mapping_Nan_movies)
```

```python
print(star_wars["seen1"].value_counts())
print(star_wars["seen2"].value_counts())
print(star_wars["seen3"].value_counts())
print(star_wars["seen4"].value_counts())
print(star_wars["seen5"].value_counts())
print(star_wars["seen6"].value_counts())
```

```python
movie_names = ["Star Wars: Episode I  The Phantom Menace", 
               "Star Wars: Episode II  Attack of the Clones", 
              "Star Wars: Episode III  Revenge of the Sith", 
              "Star Wars: Episode IV  A New Hope", 
               "Star Wars: Episode V The Empire Strikes Back", 
               "Star Wars: Episode VI Return of the Jedi"]
```

```python
#star_wars["seen1"] = star_wars["seen1"].map({"Star Wars: Episode I  The Phantom Menace":True, np.NaN:False})
for i,movie in enumerate(movie_names,1):
    mapping = {movie:True, np.NaN: False}
    star_wars["seen"+str(i)] = star_wars["seen"+str(i)].map(mapping)
    
```

```python
star_wars.head(5)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
                                "Unnamed: 10":"ranking_2", 
                                "Unnamed: 11":"ranking_3", 
                                "Unnamed: 12":"ranking_4",
                                "Unnamed: 13":"ranking_5", 
                               "Unnamed: 14":"ranking_6"})
```

```python
movie_list = ["ranking_"+str(i) for i in range(1,6)]

mean_movie_ranks = star_wars[movie_list].mean()
```

```python
%matplotlib inline
mean_movie_ranks.plot.bar()
```

We've cleaned up the data, converted Yes and No to boolean values, and summarized the rankings of the Star Wars movies (lower ranks are better!). Episodes 2 and 3 are the lowest ranked (no surprise!).

```python
seen_movies_list = ["seen" + str(i) for i in range (1,6)]
star_wars[seen_movies_list].sum().plot.bar()
```

Number that have seen each movie

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
male_mean_movie_ranks = males[movie_list].mean()
female_mean_movie_ranks = females[movie_list].mean()
male_mean_movie_ranks.plot.bar()
female_mean_movie_ranks = females[movie_list].mean()
```

```python
female_mean_movie_ranks.plot.bar()
```

Males and females have roughly similar preferences, though females seem to like episode 1 less.
